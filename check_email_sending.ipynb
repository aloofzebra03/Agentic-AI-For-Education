```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import dotenv
import os

dotenv.load_dotenv(dotenv_path=".env",override=True)

# ----------------------------
# SMTP CONFIG (dnspark / cPanel)
# ----------------------------
SMTP_SERVER = "cp1.dnspark.in"
SMTP_PORT = 587  # STARTTLS

SENDER_EMAIL = os.getenv("SENDER_EMAIL")
SENDER_PASSWORD = os.getenv("SENDER_PASSWORD")

RECIPIENTS = [
    "aryansinghalaattss@gmail.com",
    "mail2anuragmn@gmail.com"
]

# ----------------------------
# Email Content
# ----------------------------
SUBJECT = "Test for Python Email Code"
BODY = """Hello,

This is a test email sent using Python SMTP from Padaams cp1 domain-based email.

Regards,
Aryan
"""

# ----------------------------
# Build Email
# ----------------------------
msg = MIMEMultipart()
msg["From"] = SENDER_EMAIL
msg["To"] = ", ".join(RECIPIENTS)
msg["Subject"] = SUBJECT

msg.attach(MIMEText(BODY, "plain"))

# ----------------------------
# Send Email
# ----------------------------
try:
    server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
    server.ehlo()
    server.starttls()
    server.login(SENDER_EMAIL, SENDER_PASSWORD)

    server.sendmail(SENDER_EMAIL, RECIPIENTS, msg.as_string())
    server.quit()

    print("✅ Email sent successfully!")

except Exception as e:
    print("❌ Failed to send email:")
    print(e)
```

```python
import smtplib
import ssl
import os
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from dotenv import load_dotenv

load_dotenv()

SMTP_SERVER = "cp1.dnspark.in"
SMTP_PORT = 587  # STARTTLS

SENDER_EMAIL = os.getenv("SENDER_EMAIL")
SENDER_PASSWORD = os.getenv("SENDER_PASSWORD")

RECIPIENTS = [
    "aryansinghalaattss@gmail.com",
    "mail2anuragmn@gmail.com"
]

print("USER:", SENDER_EMAIL)
print("PASS LENGTH:", len(SENDER_PASSWORD) if SENDER_PASSWORD else None)

msg = MIMEMultipart()
msg["From"] = SENDER_EMAIL
msg["To"] = ", ".join(RECIPIENTS)
msg["Subject"] = "Python SMTP test via 587 (STARTTLS)"

msg.attach(MIMEText(
    "Hello,\n\nThis email is sent using Python SMTP over port 587.\n\nRegards,\nAryan",
    "plain"
))

try:
    context = ssl.create_default_context()

    server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT, timeout=30)
    server.set_debuglevel(1)  # VERY IMPORTANT: shows SMTP conversation

    server.ehlo()             # EHLO before TLS
    server.starttls(context=context)
    server.ehlo()             # EHLO AGAIN after TLS (critical)

    server.login(SENDER_EMAIL, SENDER_PASSWORD)
    server.sendmail(SENDER_EMAIL, RECIPIENTS, msg.as_string())
    server.quit()

    print("✅ Email sent successfully via 587!")

except Exception as e:
    print("❌ Failed to send email:")
    print(e)
```
